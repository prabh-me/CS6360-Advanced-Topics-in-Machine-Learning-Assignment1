```python
#Define k-fold cross validation method
def k_fold_cross_valid(data, labels, classifier):
    (row_count, columns_count) = np.shape(data)
    num_folds = sklearn.cross_validation.KFold(row_count, n_folds=10)

    for train, test in num_folds:
        train_data = data[train]
        test_data = data[test]
        labels_train = labels[train]
        labels_test = labels[test]

    classifier.fit(train_data, labels_train)
    return classifier
```
